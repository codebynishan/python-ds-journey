```python
import pandas as pd
```

## Merging data 

```python
d1 = pd.DataFrame(
    {
        "key":["a","b","c","c","d","e"],
        "num1":range(6)
    }
)

d2 = pd.DataFrame(
    {
        "key":["b","c","e","f"],
        "num2":range(4)
    }
)
```

```python
pd.merge(d1,d2)
```

```python
d3= pd.DataFrame(
    {
        "key1":["a","b","c","c","d","e"],
        "num1":range(6)
    }
)

d4 = pd.DataFrame(
    {
        "key2":["b","c","e","f"],
        "num2":range(4)
    }
)
```

```python
d3
```

```python
d4
```

```python
pd.merge(d3,d4,left_on="key1",right_on="key2")
```

## outer 

```python
d1
```

```python
d2
```

```python
pd.merge(d1,d2,how="outer")
```

## left join

```python
pd.merge(d1,d2,how="left")
```

## right join

```python
pd.merge(d1,d2,how="right")
```

```python
df5=pd.DataFrame(
    {
        "letter":["a","a","b","b","a","c"],
        "num":range(6)
    }
)

df6 =pd.DataFrame(
    {
        "value":[3,5,7]},
    index=["a","b","e"]
    
)
```

```python
df5
```

```python
df6
```

```python
pd.merge(df5,df6,left_on="letter",
         right_index=True)
```

```python
data1 = pd.DataFrame(

    [[1,2],[3,4],[5,6]],

    index=["a","c","d"],

    columns=["Tom","Tim"])
 
data2 = pd.DataFrame(

    [[7,8],[9,10],[11,12],[13,14]],

    index=["a","b","e","f"],

    columns=["Sam","Kim"])
 
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2,right_index=True,
         left_index=True,
         how ='outer')
```

```python
data1.join(data2)
```

```python
data1.join(data2,how='outer')
```
